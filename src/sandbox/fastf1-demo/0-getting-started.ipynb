## Plotting Telemetry for a Specific Driver

```python
session = fastf1.get_session(2023, 'Bahrain', 'Q')
session.load()
```

```python
# Loading data for Leclerc
fast_leclerc = session.laps.pick_driver('LEC').pick_fastest()
lec_car_data = fast_leclerc.get_car_data()

lec_car_data
```

```python
fastf1.plotting.setup_mpl()

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(18, 9), sharex=True, constrained_layout=True)
axes[0].plot(lec_car_data['Time'], lec_car_data['Speed'], label='Speed')
axes[1].plot(lec_car_data['Time'], lec_car_data['Throttle'], label='Throttle')
axes[1].plot(lec_car_data['Time'], lec_car_data['Brake'] * 100, label='Brake')
axes[2].plot(lec_car_data['Time'], lec_car_data['nGear'], label='Gear')
axes[0].set_xlabel('')
axes[1].set_xlabel('')
axes[2].set_xlabel('Time')
axes[0].set_ylabel('Speed (km/h)')
axes[1].set_ylabel('Throttle (%) | Brake (On/Off)')
axes[2].set_ylabel('Gear')
fig.suptitle("Charles Leclerc's fastest lap at Bahrain 2023")
for ax in axes:
    ax.legend()
```
